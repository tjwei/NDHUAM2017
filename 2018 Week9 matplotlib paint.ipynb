```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy import sin, cos, pi
```

```python
plt.figure(figsize=(7,7))
plt.xlim(-10, 10)
plt.ylim(-10, 10)
t = np.linspace(0, 2*pi, 100)
plt.plot(9*cos(t), 9*sin(t), 'r')
plt.plot(8*cos(t), 8*sin(t), 'r')
plt.plot(4.7*cos(t), 4.7*sin(t), 'r')
plt.plot(5.6*cos(t), 5.6*sin(t), 'r')
t = np.linspace(0, 2*pi, 4+1)
plt.plot(8*cos(t), 8*sin(t), 'r', linewidth=4)
plt.plot(8*cos(t+pi/4), 8*sin(t+pi/4), 'r', linewidth=4)

t = np.linspace(0, 2*pi, 1000)
r = 8.5+0.3*sin(t*170)+0.2*sin(t*370)+0.1*sin(t*50)
plt.plot(r*cos(t), r*sin(t), 'r')
r = 8.5+0.3*sin(t*666)+0.2*sin(t*444)+0.1*sin(t*150)
plt.plot(r*cos(t), r*sin(t), 'r')

t = np.linspace(0, 2*pi, 1000)
s = np.linspace(0, 150, 1000)
r = 5.15+ 0.7*(s%1-0.5)
plt.plot(r*cos(t), r*sin(t), 'r')
```

```python
plt.figure(figsize=(7,7))
plt.xlim(-10, 10)
plt.ylim(-10, 10)
axes = plt.gca()
axes.set_facecolor((.7,.8, 1))


t = np.linspace(0, pi, 100)
r = 9
for color in [(1,0,0), (1, .6, 0), (1,1,0), (0,1,0),
             (0.3,0.3,1), (0.5,0.5,0.9), (1,0.3,1)]:
    plt.plot(r*cos(t), r*sin(t)-1.5, color=color, linewidth=6)
    #plt.plot(t*20/(pi)-10, 2*sin(-2*t)+r-6.5, color=color, linewidth=6)
    r-=0.3
t = np.linspace(0, 2*pi, 100)
plt.fill(-7.5+1.2*cos(t),7.5+1.2*sin(t), color=(1,1,0.6))
t = np.linspace(0, 8*pi, 12)
plt.fill(-7.5+1.5*cos(t),7.5+1.5*sin(t), color=(1,1,0.6))

x = np.linspace(-10,10, 100)
y=-7*np.ones_like(x)
plt.plot(x, y, color=(0,1,0), linewidth=200)
y = -np.abs(x+2.5)-1.5
plt.plot(x,y, linewidth=200, color=(0,1,0))

t = np.linspace(0, pi, 100)
plt.plot(t*20/(pi)-10, 1.2*sin(-2*t)-5.5, color=(0.2,0.4,1), linewidth=26)


t = np.linspace(0, 2*pi, 100)
plt.fill(5.5+3.5*cos(t),7+1.5*sin(t), color=(.95,.95,.95))
plt.fill(4.3+1.5*cos(t),7.7+1.4*sin(t), color=(.95,.95,.95))
plt.fill(6.3+1.8*cos(t),7.9+1.4*sin(t), color=(.95,.95,.95))
plt.fill(8.8+0.7*cos(t),7.0+1.1*sin(t), color=(.95,.95,.95))
```
