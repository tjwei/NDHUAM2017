```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy import sin, cos, pi
```

```python
plt.figure(figsize=(7,7))
plt.xlim(-10, 10)
plt.ylim(-10, 10)
t = np.linspace(0, 2*pi, 100)
plt.plot(9*cos(t), 9*sin(t), 'r')
plt.plot(8*cos(t), 8*sin(t), 'r')
plt.plot(4.7*cos(t), 4.7*sin(t), 'r')
plt.plot(5.6*cos(t), 5.6*sin(t), 'r')
t = np.linspace(0, 2*pi, 4+1)
plt.plot(8*cos(t), 8*sin(t), 'r', linewidth=4)
plt.plot(8*cos(t+pi/4), 8*sin(t+pi/4), 'r', linewidth=4)

t = np.linspace(0, 2*pi, 1000)
r = 8.5+0.3*sin(t*170)+0.2*sin(t*370)+0.1*sin(t*50)
plt.plot(r*cos(t), r*sin(t), 'r')
r = 8.5+0.3*sin(t*666)+0.2*sin(t*444)+0.1*sin(t*150)
plt.plot(r*cos(t), r*sin(t), 'r')

t = np.linspace(0, 2*pi, 1000)
s = np.linspace(0, 150, 1000)
r = 5.15+ 0.7*(s%1-0.5)
plt.plot(r*cos(t), r*sin(t), 'r')
```
