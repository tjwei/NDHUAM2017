```
ftable_full=np.logspace(0,np.log10(2),13)*262
ftable = ftable_full[[0,2,4,5,7,9,11]]
```

```
import time
```

```
for i in range(7):
  y = np.sin(t * ftable[i] * 2* np.pi)
  display(Audio(data=y, rate=rate, autoplay=True))
  time.sleep(1.1)
```
