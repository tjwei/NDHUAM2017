```

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```
#@title 默认标题文本 { vertical-output: true }
plt.figure(figsize=(5,5))

t = np.linspace(0, 2*np.pi, 101)
plt.plot(np.cos(t), np.sin(t), 'r--')
plt.plot(.7*np.cos(t), .7*np.sin(t), 'r')
plt.plot(1.3*np.cos(t), 1.3*np.sin(t), 'r')

for i in range(8):
    t0 = 2*np.pi/8*i
    x0, y0 = 1.15*np.cos(t0), 1.15*np.sin(t0)
    plt.plot(.15*np.cos(t)+x0, .15*np.sin(t)+y0, 'r')
    
t = np.linspace(0, 4*np.pi, 6)
plt.plot(.7*np.cos(t+np.pi/10), 
         .7*np.sin(t+np.pi/10), 'r')

t = np.linspace(0, 2*np.pi, 5)
plt.plot(np.cos(t), np.sin(t), 'r:')
plt.plot(np.cos(t+np.pi/4), np.sin(t+np.pi/4), 'r:')

t = np.linspace(0, 2*np.pi, 9)
plt.plot(1.15*np.cos(t), 1.15*np.sin(t), 'rp')

#plt.savefig('circle.png')
```

```
from scipy.ndimage import gaussian_filter
def blur(img, dpi):
    return gaussian_filter(img, (3,3,0)), -0, 0
def plotx(*a, linewidth=2, **k):
    plt.plot(*a, **k, linewidth=15*linewidth, alpha=0.1, agg_filter=blur)
    plt.plot(*a, **k, linewidth=7*linewidth, alpha=0.3, agg_filter=blur)
    plt.plot(*a, **k, linewidth=linewidth)
```

```
from scipy.ndimage import gaussian_filter
def Blur(sigma):
    def blur(img, dpi):
        return gaussian_filter(img, (sigma,sigma,0)), -0, 0
    return blur
```
